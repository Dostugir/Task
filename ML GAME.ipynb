Guessing Game


```python

from random import randint
R_num= randint(1,100)
count = 0
user=int(input("Input a number 1 to 100:"))
count+=1
while R_num!=user:
    if user>R_num:
        print("Too High!")
        user=int(input("Try again! \n Input a number 1 to 100:"))
        count+=1
    elif user<R_num:
        print("Too Low!")
        user=int(input("Try again! \n Input a number 1 to 100:"))
        count+=1
    elif user==R_num:
        break
print("Finally matched.\n You tried ",count," Times","\n AI chosen ",R_num)
```

Rock Paper scissor


```python
import random
R_list = ["rock","paper","scissor"]
while True:
    AI_choice=random.choice(R_list)
    user = input("Input [rock,paper,scissor] choice according to given value:")
    user = user.lower()
    if AI_choice==user:
        print("You Win!")
        print("If want to quit this game Enter 'exit'")
    elif user=="exit":
        break
    else:
        print("You Loose!")
        print("If want to quit this game Enter 'exit'")
```
